```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Rastrigin function
def rastrigin(x):
    n = len(x)
    return 10 * n + sum([xi**2 - 10 * np.cos(2 * np.pi * xi) for xi in x])

# Random Hill Climbing Algorithm
def random_hill_climbing(func, n_dim, bounds, max_iter=1000, step_size=0.1):
    # Initialize a random solution within bounds
    current_solution = np.random.uniform(bounds[0], bounds[1], n_dim)
    current_value = func(current_solution)

    best_solution = current_solution
    best_value = current_value

    history = []  # To store the progress

    for i in range(max_iter):
        # Generate a random neighbor
        neighbor = current_solution + np.random.uniform(-step_size, step_size, n_dim)
        # Ensure the neighbor is within bounds
        neighbor = np.clip(neighbor, bounds[0], bounds[1])

        # Evaluate the neighbor
        neighbor_value = func(neighbor)

        # If the neighbor is better, move to the neighbor
        if neighbor_value < current_value:
            current_solution = neighbor
            current_value = neighbor_value
            print(f"{current_solution}")
            print(f"{current_value}")


            # Update the best solution if it's better
            if neighbor_value < best_value:
                best_solution = neighbor
                best_value = neighbor_value

        # Record progress
        history.append(current_value)

    return best_solution, best_value, history

# Parameters
n_dim = 2                 # Number of dimensions
bounds = (-5.12, 5.12)    # Bounds for Rastrigin function
max_iter = 1000           # Maximum iterations
step_size = 0.1           # Step size for neighbors

# Run the optimization
best_solution, best_value, history = random_hill_climbing(rastrigin, n_dim, bounds, max_iter, step_size)

# Print results
print("Best Solution:", best_solution)
print("Best Value:", best_value)

# Plot the convergence
plt.figure(figsize=(10, 6))
plt.plot(history)
plt.title('Convergence of Random Hill Climbing on Rastrigin Function')
plt.xlabel('Iteration')
plt.ylabel('Function Value')
plt.grid()
plt.show()
```
