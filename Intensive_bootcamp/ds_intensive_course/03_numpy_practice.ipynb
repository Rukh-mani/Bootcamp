```python
#v3 = np.array([1, -2])

def plot_vectors(vectors, colors, alpha=1):
    """
    Plot set of vectors.

    Parameters:
    - vectors: list of vectors to be plotted.
    - colors: colors for the vectors.
    """
    plt.figure()
    plt.axvline(x=0, color='grey', lw=1)
    plt.axhline(y=0, color='grey', lw=1)
    for i in range(len(vectors)):
        if isinstance(alpha, list):
            alpha_i = alpha[i]
        else:
            alpha_i = alpha
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=colors[i], alpha=alpha_i)
    plt.grid()
    plt.xlim(-5, 5)
    plt.ylim(-5, 5)
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.title('Vector Visualization')
    plt.show()

v1 = np.array([1, 0])
v2 = np.array([0, 1])
v3 = v1 + v2


# Plot the vectors
plot_vectors([v1, v2, v3], ['blue', 'red', 'green'])

"""
plt.figure()
pt1 = [1,2,3]
pt2 = [0,1,2]
plt.scatter(pt1,pt2)
plt.show()
"""
#print(np.dot(v1, v2))
#v4 = np.array([0,1,2,3])
#len(v4)

print(v3)
```

### Euclidean distance
$||\vec{a}-\vec{b}|| = \sqrt{(a_1-b_1)^2 + (a_2-b_2)^2 + \cdots + (a_n-b_n)^2}$

```python
def euclidean_distance(p, q):
    """
    Compute the Euclidean distance between two data points.

    Parameters:
    - p, q: numpy arrays or lists representing the data points.

    Returns:
    - Euclidean distance between p and q.
    """
    p = np.array(p)
    q = np.array(q)
    return np.sqrt(np.sum((p - q)**2))

euclidean_distance(p, q)
```

```python
from scipy.spatial.distance import pdist, squareform

# Calculate pairwise Euclidean distances between rows (friends/you)
distances = squareform(pdist(data, 'euclidean'))
print("Pairwise distances:\n", distances)

np.fill_diagonal(distances, np.inf)  # Fill diagonal with infinity to ignore self-comparisons
individuals_indices = np.unravel_index(np.argmin(distances), distances.shape)

individuals_map = ["You", "Friend1", "Friend2", "Friend3"]

print(f"{individuals_map[individuals_indices[0]]} and {individuals_map[individuals_indices[1]]} have the most similar tastes.")

```
