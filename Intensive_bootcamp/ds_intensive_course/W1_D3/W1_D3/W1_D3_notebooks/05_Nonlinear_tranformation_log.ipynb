```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 1: Generate synthetic data with an exponential relationship
np.random.seed(0)
x = np.linspace(1, 10, 100)  # Feature x
y = np.exp(0.5 * x) + np.random.normal(scale=5, size=x.shape)  # Exponential relationship with noise

# Ensure all y values are positive by adding a small constant (if necessary)
y = np.where(y <= 0, 0.01, y)  # Replace any non-positive values with a small positive value

# Plot the original exponential data
plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='blue', label='Original Exponential Data')
plt.title("Exponential Relationship Between x and y")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# Step 2: Apply a logarithmic transformation to linearize the relationship
y_log = np.log(y)

# Plot the transformed data
plt.figure(figsize=(10, 5))
plt.scatter(x, y_log, color='green', label='Log-transformed Data')
plt.title("Log-transformed Data to Linearize the Relationship")
plt.xlabel("x")
plt.ylabel("log(y)")
plt.legend()
plt.show()

# Step 3: Fit a linear model to the transformed data
model = LinearRegression()
x_reshaped = x.reshape(-1, 1)
model.fit(x_reshaped, y_log)
y_log_pred = model.predict(x_reshaped)

# Plot the linear fit on the transformed data
plt.figure(figsize=(10, 5))
plt.scatter(x, y_log, color='green', label='Log-transformed Data')
plt.plot(x, y_log_pred, color='red', label='Linear Fit on Log-transformed Data')
plt.title("Linear Fit on Log-transformed Data")
plt.xlabel("x")
plt.ylabel("log(y)")
plt.legend()
plt.show()

# Step 4: Print the slope and intercept of the linear model
print("Slope:", model.coef_[0])
print("Intercept:", model.intercept_)
```
