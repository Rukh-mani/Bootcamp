




*   Two categories were considered: **Romance** and **Action**.
*   Four individuals were asked about their preferences between these two categories.

*   Each person's preference was represented as a vector, with the two components being the levels of Romance and Action they prefer.
*   The four vectors are:

    

      Person 1.   (100, 0): Strong preference for Romance, no preference for Action.
      Person 2.   (90, 10): High preference for Romance, slight preference for Action.
      Person 3.   (10, 90): High preference for Action, slight preference for Romance.
      Person 4.   (0, 100): Strong preference for Action, no preference for Romance.

These vectors were then plotted on a 2D graph with Romance on the x-axis and Action on the y-axis to visually represent their preferences.
```
# This is formatted as code
```



````python
import plotly.graph_objects as go
import numpy as np

# Define the vectors
vectors = np.array([[100, 0], [90, 10], [10, 90], [0, 100]])

# Create the figure
fig = go.Figure()

# Add vectors to the plot
for i, vec in enumerate(vectors):
    fig.add_trace(go.Scatter(
        x=[0, vec[0]],
        y=[0, vec[1]],
        mode="lines+markers+text",
        text=f"Person {i+1}",
        name=f"Person {i+1}",
        textposition="top center",
        textfont=dict(size=20, family='Arial Bold')
    ))

# Set the axis labels
# fig.update_layout(
#     title="Vectors on Romance-Action Plane",
#     xaxis_title="Romance",
#     yaxis_title="Action",

#     xaxis=dict(range=[0, 110]),
#     yaxis=dict(range=[0, 110])
# )
font_size = 25
fig.update_layout(
    title="Vectors on Romance-Action Plane",
    title_font=dict(size=font_size, family='Arial Bold', color='black'),
    xaxis_title="Romance",
    yaxis_title="Action",
        width=800,  # 10 cm ~ 394 pixels
    height=800,
    xaxis_title_font=dict(size=font_size, family='Arial Bold', color='black'),
    yaxis_title_font=dict(size=font_size, family='Arial Bold', color='black'),
    xaxis=dict(range=[-20, 110], tickfont=dict(size=font_size, family='Arial Bold')),
    yaxis=dict(range=[0, 110], tickfont=dict(size=font_size, family='Arial Bold')),
    font=dict(size=14, family='Arial Bold')
)
````

The preference of a person for a genre can be easily understood from the score.
The score also reveals which among 4 have similar preference for a genre.

It can also be found from the angle between the two vectors.
If the angle between two vectors is small, then it means they are closely related.
If the angle between two vectors is 90 degree, it means they is not relation.
If the angle is 180 degree, it means they are opposite relation.
