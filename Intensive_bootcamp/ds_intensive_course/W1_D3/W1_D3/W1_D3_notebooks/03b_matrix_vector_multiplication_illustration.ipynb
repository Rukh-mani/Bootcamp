```python
import numpy as np
import matplotlib.pyplot as plt

# Define a vector
vector = np.array([2, 3])

# Plot the original vector
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='blue')
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)
plt.grid()
plt.title("Original Vector")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Define a matrix for transformation
matrix = np.array([[1, 2],
                   [3, 1]])

# Perform matrix-vector multiplication
transformed_vector = matrix @ vector

# Plot the transformed vector
plt.figure(figsize=(6, 6))
# Plot original vector for reference
plt.quiver(0, 0, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='blue', label="Original Vector")
# Plot transformed vector
plt.quiver(0, 0, transformed_vector[0], transformed_vector[1], angles='xy', scale_units='xy', scale=1, color='red', label="Transformed Vector")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)
plt.grid()
plt.legend()
plt.title("Transformed Vector after Matrix Multiplication")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()

# Display the results
print("Original Vector:", vector)
print("Matrix:\n", matrix)
print("Transformed Vector:", transformed_vector)
```
