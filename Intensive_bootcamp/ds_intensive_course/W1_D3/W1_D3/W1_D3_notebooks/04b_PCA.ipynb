```python
from sklearn.datasets import load_iris
import numpy as np
import matplotlib.pyplot as plt

# Load the iris dataset
iris = load_iris()
data = iris.data

# Compute the covariance matrix
cov_matrix = np.cov(data, rowvar=False)

# Perform eigen decomposition on the covariance matrix
eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

# Sort the eigenvalues and eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[sorted_indices]
eigenvectors = eigenvectors[:, sorted_indices]

# Select the first two principal vectors (eigenvectors)
principal_vectors = eigenvectors[:, :2]

# Project the data onto the first two principal components
projected_data = data.dot(principal_vectors)

# Plot the data in the new principal component space
plt.figure(figsize=(10, 7))
plt.scatter(projected_data[:, 0], projected_data[:, 1], c=iris.target, cmap='viridis', edgecolor='k', s=50)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Iris Data Projected onto First Two Principal Components")
plt.colorbar(label='Target Class')
plt.show()
```

```python
from sklearn.decomposition import PCA



# Apply PCA using sklearn to confirm the results

pca = PCA(n_components=2)

pca_data = pca.fit_transform(data)



# Plot the PCA-transformed data

plt.figure(figsize=(10, 7))

plt.scatter(pca_data[:, 0], pca_data[:, 1], c=iris.target, cmap='viridis', edgecolor='k', s=50)

plt.xlabel("Principal Component 1")

plt.ylabel("Principal Component 2")

plt.title("Iris Data Projected onto First Two Principal Components (using sklearn PCA)")

plt.colorbar(label='Target Class')

plt.show()
```
