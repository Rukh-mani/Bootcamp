```python
#TODO separate customer IDs to groups A and B randomly. Hint - use random module

shuffled_df = df.sample(frac=1, ignore_index=True)
groupA = shuffled_df.iloc[:int(len(shuffled_df)/2)] #all upto nth row
groupB = shuffled_df.iloc[int(len(shuffled_df)/2):] #from n to remaining


```

```python
#TODO use at least 2 data attributes to check that your random shuffling results into two subsets with equal medians
#Hint use https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.kruskal.html

scipy.stats.kruskal(groupA.Income, groupB.Income, nan_policy='omit') #omitting nans is alright since we are checking the median
```
