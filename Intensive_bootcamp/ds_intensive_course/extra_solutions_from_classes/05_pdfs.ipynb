```python
# Import libraries
from scipy.stats import norm,expon
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
####################################################
### Exercise with height of kids

##########################
## Without standardization
print('Without standardization')

my_loc = 130
my_scale = np.sqrt(36)

# Taller than 136cm
prob = 1-norm.cdf(136, loc=my_loc, scale=my_scale)
print(f'{prob:.2f}')

# Smaller than 118cm
prob = norm.cdf(118, loc=my_loc, scale=my_scale)
print(f'{prob:.2f}')

#Between 133 and 127 cm
p1 = norm.cdf(133, loc=my_loc, scale=my_scale)
p2 = norm.cdf(127, loc=my_loc, scale=my_scale)
print(f'{p1-p2:.2f}')

##########################
## With standardization
print('With standardization')

#Question 1
prob_1 = 1-norm.cdf(1, loc=0, scale=1)
print(f'{prob_1:.2f}')

#Question 2
prob_2 = norm.cdf(-2, loc=0, scale=1)
print(f'{prob_2:.2f}')

#Question 3
prob_3 = norm.cdf(0.5, loc=0, scale=1) - norm.cdf(-0.5, loc=0, scale=1)
print(f'{prob_3:.2f}')
```

```python
# Generate 1000 random numbers from an exponential distribution
random_numbers = [-np.log(1 - random.random()) / 0.5 for _ in range(1000)]
random_numbers = expon.rvs(loc = 5, size = 100000)

# Calculate mean and variance
mean = np.mean(random_numbers) #TODO use np.mean to calculate mean
variance = np.var(random_numbers) #TODO use np.mean to calculate variance

# Plot a histogram
plt.hist(random_numbers, bins=20, edgecolor='k')
plt.title(f"Exponential Distribution\nMean: {mean:.2f}, Variance: {variance:.2f}")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
