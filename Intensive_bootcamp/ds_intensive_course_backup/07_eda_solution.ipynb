```python
import pandas as pd

# Load the data
train_data = pd.read_csv('./data/titanic.csv')

# Display the first few rows of the dataframe
train_head = train_data.head()

# Get the summary statistics of the training data
train_description = train_data.describe()

train_head, train_description
```

```python
# Check for missing values in each column
missing_values = train_data.isnull().sum()

# Check data types of each column
data_types = train_data.dtypes

missing_values, data_types
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# List of columns for univariate analysis
columns_to_analyze = ['Survived', 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']

#
sns.countplot(data=train_data, x="Survived", palette='viridis')
plt.show()
sns.histplot(train_data["Age"], bins=30, kde=True, color='blue')
plt.show()
```

```python
# Create a figure to hold multiple subplots
plt.figure(figsize=(20, 15))

# Create subplots for each column
for i, column in enumerate(columns_to_analyze, 1):
    plt.subplot(3, 3, i)
    
    # If the column is non-numeric, use a count plot
    if train_data[column].dtype == 'object' or column in ['Survived', 'Pclass']:
        sns.countplot(data=train_data, x=column, palette='viridis')
        plt.title(f'Distribution of {column}')
    # For numeric data, use a histogram
    else:
        sns.histplot(train_data[column], bins=30, kde=True, color='blue')
        plt.title(f'Distribution of {column}')
    
    plt.tight_layout()

# Show the plots
plt.show()
```

```python
# Create a figure to hold the subplots
plt.figure(figsize=(20, 7))

# Survival rate by Sex
plt.subplot(1, 2, 1)
sns.countplot(data=train_data, x='Sex', hue='Survived', palette='viridis')
plt.title('Survival Rate by Sex')

# Survival rate by Pclass
plt.subplot(1, 2, 2)
sns.countplot(data=train_data, x='Pclass', hue='Survived', palette='viridis')
plt.title('Survival Rate by Pclass')

plt.tight_layout()
plt.show()
```

```python
# Create a figure to hold the subplots
plt.figure(figsize=(20, 7))

# Box plot of Age versus Survival
plt.subplot(1, 2, 1)
sns.boxplot(data=train_data, x='Survived', y='Age', palette='viridis')
plt.title('Age Distribution by Survival')

# Box plot of Fare versus Survival
plt.subplot(1, 2, 2)
sns.boxplot(data=train_data, x='Survived', y='Fare', palette='viridis')
plt.title('Fare Distribution by Survival')

plt.tight_layout()
plt.show()

# Scatter plot of Age versus Fare, colored by Survival
plt.figure(figsize=(10, 7))
sns.scatterplot(data=train_data, x='Age', y='Fare', hue='Survived', style='Survived', palette='viridis')
plt.title('Survival by Age and Fare')
plt.show()
```
