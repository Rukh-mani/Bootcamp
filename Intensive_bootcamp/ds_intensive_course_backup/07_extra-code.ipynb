```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import numpy as np

# Generate some dummy data
np.random.seed(0)
data = np.random.randn(1000)

# Using matplotlib
plt.hist(data, bins=30, alpha=0.5, color='blue')
plt.title('Histogram (matplotlib)')
plt.xlabel('Data points')
plt.ylabel('Frequency')
plt.show()

# Using seaborn
sns.histplot(data, bins=30, kde=False, color='red')
plt.title('Histogram (seaborn)')
plt.xlabel('Data points')
plt.ylabel('Frequency')
plt.show()

# Using plotly
fig = px.histogram(data, nbins=30, title='Histogram (plotly)')
fig.update_layout(xaxis_title='Data points', yaxis_title='Frequency')
fig.show()
```

```python
import plotly.express as px
from sklearn import datasets
from sklearn.decomposition import PCA

# Load the iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Perform PCA to reduce the data to 3 dimensions for visualization
pca = PCA(n_components=3)
components = pca.fit_transform(X)

# Create a 3D scatter plot using Plotly
fig = px.scatter_3d(
    x=components[:, 0], 
    y=components[:, 1], 
    z=components[:, 2], 
    color=y.astype(str),  # Convert target labels to strings for color mapping
    labels={'0': 'PC 1', '1': 'PC 2', '2': 'PC 3'},
    title="3D PCA of Iris Dataset"
)

# The `show()` function will render the plot. In an interactive Python environment,
# such as Jupyter Notebook, the plot will display inline.
# In a script, this will open the plot in a browser window.
fig.show()
```
